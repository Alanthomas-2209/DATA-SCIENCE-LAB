```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd

file_path = '/content/drive/My Drive/Data_science lab/6.Outliers/weightdata.csv'
df = pd.read_csv(file_path)

def fnctn(data, col):
    mean = data[col].quantile(0.50)
    sd = data[col].std()  # Calculate the standard deviation for the specific column
    lower_limit = mean - 3 * sd
    upper_limit = mean + 3 * sd
    return data[(data[col] >= lower_limit) & (data[col] <= upper_limit)]

columns = ['Weight(Pounds)']

for column in columns:
    df = fnctn(df, column)

output = '/content/drive/My Drive/Data_science lab/6.Outliers/Output_sd-Sheet1.csv'
df.to_csv(output, index=False)

drive.mount("/content/drive", force_remount=True)
```

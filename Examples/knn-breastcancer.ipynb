```python
from sklearn.preprocessing import StandardScaler
```

```python
input_scaled = StandardScaler().fit_transform(inputs)
```

```python
input_scaled
```

```python
df_scaled = pd.DataFrame(input_scaled, index=inputs.index, columns=inputs.columns)
```

```python
df_scaled.describe()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
             df_scaled, target, test_size = 0.2, random_state=42)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
  
knn.fit(X_train, y_train)
```

```python
print(knn.score(X_test, y_test))
```

```python
p=knn.predict(X_test)
```

```python
from sklearn import metrics

```

```python
confusion_matrix = metrics.confusion_matrix(y_test, p)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

```

```python
import matplotlib.pyplot as plt

```

```python
cm_display.plot()
plt.show()

```
